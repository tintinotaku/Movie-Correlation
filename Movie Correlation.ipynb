```python
# import libraries and dataset

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)

# import dataset
df = pd.read_csv('E:\PROJECT\Python Project\Movie\movies.csv')
df.head(5)
```

```python
#Total columns and rows
df.shape
```

```python
#Check for the missing data
for col in df.columns:
    a = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,a*100))
```

```python
#Drop rows that contain Null values
df = df.dropna()
df.head()
```

```python
#DOUBLE CHECK for the missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing*100))
```

```python
#Check data type
df.dtypes
```

```python
#Change column type ('votes', 'budget', 'gross','runtime' into int)
df['votes'] = df['votes'].astype(int)
df['budget'] = df['budget'].astype(int)
df['gross'] = df['gross'].astype(int)
df['runtime'] = df['runtime'].astype(int)
df.head()
```

```python
#Create a new "year" column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
#drop the old "year" column

df.drop('year', axis = 1, inplace = True)
```

```python
#sort value of gross
df.sort_values(by='gross',ascending = False, inplace = False)
df.head()
```

```python
#Drop any duplicate value for all entire columns (Actually it's distinct count in SQL)
df = df.drop_duplicates()
df.head()
```

```python
df.shape
```

```python
#Budget  high correlation
#Company high correlation

#Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.xlabel('Budget',fontsize=15)
plt.ylabel('Gross',fontsize=15)
plt.title("Correlation between budget & gross",fontsize=20)
plt.ticklabel_format(axis='both',style='plain')
plt.show()
```

```python
#Plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross', data =df, scatter_kws={'color':'gray'}, line_kws={'color':'lightseagreen'})
plt.ticklabel_format(axis='both',style='plain')
```

```python
#Let's start looking at correlation

df.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
#High correlation between budget and gross

correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matric for Numeric Features", fontsize=20)
plt.xlabel("Movie Features", fontsize=15)
plt.ylabel("Movie Features", fontsize = 15)
```

```python
# Bin mi ct c cc nh dng ngoi tr float, int tr thnh nh dng float, int  c th coi mi lin h ca n trong Correlation Matrix

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
#Correlation matrix for every fields 
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matric for Numeric Features", fontsize=20)
plt.xlabel("Movie Features", fontsize=15)
plt.ylabel("Movie Features", fontsize = 15)
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending = True)
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
